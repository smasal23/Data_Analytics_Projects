# Import Libraries and Modules

```python
import requests
import pandas as pd
from src.config.settings import (Coingecko_Base_URL, Coingecko_Market_Endpoint, Coingecko_VS_Currency, Coingecko_Per_Page, Coingecko_Order)
```

# Call CoinGecko markets API

```python
url = f"{Coingecko_Base_URL}{Coingecko_Market_Endpoint}"
```

```python
parameters = {
    "vs_currency" : Coingecko_VS_Currency,
    "order" : Coingecko_Order,
    "per_page" : Coingecko_Per_Page,
    "page" : 1,
    "sparkline" : False
    }
```

# Store raw JSON response

```python
CG_markets_1 = []
for page in range(1, 4):
    parameters["page"] = page
    response = requests.get(url, parameters)
    data = response.json()
    CG_markets_1.extend(data)
```

```python
CG_markets_2 = []
for page in range(4, 6):
    parameters["page"] = page
    response = requests.get(url, parameters)
    data = response.json()
    CG_markets_2.extend(data)
```

```python
# Merge the two lists while deduplicating by comprehending the dictionaries upon addition.
# CG_markets = list({i["id"] : i for i in CG_markets_1 + CG_markets_2}.values())
CG_markets = list({
    i["id"]: i
    for i in (CG_markets_1 + CG_markets_2)
    if isinstance(i, dict) and "id" in i
}.values())
```

# Convert into Dataframe

```python
df = pd.DataFrame(CG_markets)
raw_CG_markets = df.sort_values(by = "market_cap", ascending = False)
```
