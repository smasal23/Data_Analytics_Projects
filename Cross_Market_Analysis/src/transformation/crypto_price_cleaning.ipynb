```python
import pandas as pd
```

```python
df = pd.read_csv("F:\\onedrive\\Desktop\\DATA SCIENCE\\Project 1\\Cross-Market Analysis\\Cross-Market Data\\raw\\crypto\\crypto_prices.csv")
df.head()
```

# Convert timestamp â†’ date

```python
# dt.date only results into date of the datetime function.
df['date'] = pd.to_datetime(df['timestamp'], unit = 'ms').dt.date
df.head()
```

```python
df.drop(['Unnamed: 0','timestamp'], axis = 1, inplace = True)
df.head()
```

# Extract daily price

```python
# Check unique dates for each coin, nunique counts the number of unique values
unique_dates_per_coin = df.groupby('coin_id')['date'].nunique()
unique_dates_per_coin
# So, we have daily prices for all 5 coins.
```
