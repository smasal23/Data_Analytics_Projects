```python
import pandas as pd
```

```python
df1 = pd.read_csv("F:\\onedrive\\Desktop\\DATA SCIENCE\\Project 1\\Cross-Market Analysis\\Cross-Market Data\\raw\\stocks\\gspc_data.csv")

df2 = pd.read_csv("F:\\onedrive\\Desktop\\DATA SCIENCE\\Project 1\\Cross-Market Analysis\\Cross-Market Data\\raw\\stocks\\ixic_data.csv")

df3 = pd.read_csv("F:\\onedrive\\Desktop\\DATA SCIENCE\\Project 1\\Cross-Market Analysis\\Cross-Market Data\\raw\\stocks\\nsei_data.csv")
```

# Standardize OHLC

```python
df1.head()

df2.head()

df3.head()
```

```python
df1.isnull().sum()

df2.isnull().sum()

df3.isnull().sum()
```

```python
df1.info()

df2.info()

df3.info()
```

```python
# Dropping extra junk ticker row
ticker1 = df1.loc[0, "High"]
df1 = df1.drop(index = 0).reset_index(drop = True)
df1.head()

ticker2 = df2.loc[0, "High"]
df2 = df2.drop(index = 0).reset_index(drop = True)
df2.head()

ticker3 = df3.loc[0, "High"]
df3 = df3.drop(index = 0).reset_index(drop = True)
df3.head()
```

```python
# Dropping the unnecessary Price column.
df1 = df1.drop(columns = ['Price'])
df1.head()

df2 = df2.drop(columns = ['Price'])
df2.head()

df3 = df3.drop(columns = ["Price"])
df3.head()
```

# Add ticker column

```python
df1["ticker"] = ticker1
df1.head()

df2["ticker"] = ticker2
df2.head()

df3["ticker"] = ticker3
df3.head()
```

# Filter date range

```python
df1.columns
df2.columns
df3.columns
```

```python
# Renaming the Columns
df1 = df1.rename(columns = {
    "Date" : "date",
    "Close" : "close",
    "High" : "high",
    "Low" : "low",
    "Open" : "open",
    "Volume" : "volume"
    })
df1.head()

df2 = df2.rename(columns = {
    "Date" : "date",
    "Open" : "open",
    "High" : "high",
    "Low" : "low",
    "Close" : "close",
    "Volume" : "volume"
})
df2.head()

df3 = df3.rename(columns = {
    "Date" : "date",
    "Open" : "open",
    "High" : "high",
    "Low" : "low",
    "Close" : "close",
    "Volume" : "volume"
})
df3.head()
```

```python
df1["date"].dtypes
```

```python
# Fix Datatypes
# Applying the numeric date type to the columns.
df1["date"] = pd.to_datetime(df1["date"])
price_cols = ['open', 'high', 'low', 'close', 'volume']
df1[price_cols] = df1[price_cols].apply(pd.to_numeric)
df1.head()

df2["date"] = pd.to_datetime(df2["date"])
price_cols = ['open', 'high', 'low', 'close', 'volume']
df2[price_cols] = df2[price_cols].apply(pd.to_numeric)

df3["date"] = pd.to_datetime(df3["date"])
price_cols = ['open', 'high', 'low', 'close', 'volume']
df3[price_cols] = df3[price_cols].apply(pd.to_numeric)
```

```python
# Reordering the columns
cols1 = ['date', 'open', 'high', 'low', 'close', 'volume', 'ticker']
df1 = df1.reindex(columns = cols1)
df1.head()

cols2 = ["date", "open", "high", "low", "close", "volume", "ticker"]
df2 = df2.reindex(columns = cols2)
df2.head()

cols3 = ["date", "open", "high", "low", "close", "volume", "ticker"]
df3 = df3.reindex(columns = cols3)
df3.head()
```

```python
# Display the date values to full calender with NaN in missing prices. This is done so when cross-analysis with crypto takes place the data should not lie.

df1.sort_values("date")
expected_dates = pd.date_range(start = df1['date'].min(), end = df1['date'].max(), freq = 'D')
df1 = df1.set_index('date').reindex(expected_dates)

df2.sort_values("date")
expected_dates = pd.date_range(start = df2['date'].min(), end = df2['date'].max(), freq = 'D')
df2 = df2.set_index('date').reindex(expected_dates)

df3.sort_values("date")
expected_dates = pd.date_range(start = df3['date'].min(), end = df3['date'].max(), freq = 'D')
df3 = df3.set_index('date').reindex(expected_dates)
df3.head()
```

```python
# Forward Fill Prices whenever the market is closed, since the prices do not change.
price_cols = ['open', 'high', 'low', 'close', 'ticker']
df1[price_cols] = df1[price_cols].ffill()
df2[price_cols] = df2[price_cols].ffill()
df3[price_cols] = df3[price_cols].ffill()
df3.head()
```

```python
# Restore the date columns
df1 = df1.reset_index().rename(columns = {'index' : 'date'})
df1.head()

df2 = df2.reset_index().rename(columns = {'index' : 'date'})
df2.head()

df3 = df3.reset_index().rename(columns = {'index' : 'date'})
df3.head()
```

# Merge tickers

```python
# Stack all prices
df_all = pd.concat([df1, df2, df3], ignore_index = True)
df_all.head()
```
