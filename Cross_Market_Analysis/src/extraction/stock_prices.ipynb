```python
import pandas as pd
import yfinance
from src.config.settings import (Asset_Date_Ranges)
```

# Download ^GSPC

```python
stock_first = pd.to_datetime(Asset_Date_Ranges['stock']['start'], dayfirst = True)
stock_end = pd.to_datetime(Asset_Date_Ranges['stock']['end'], dayfirst = True)

gspc = yfinance.download("^GSPC", start = stock_first, end = stock_end)
```

```python
df1 = pd.DataFrame(gspc)
```

```python
df1.reset_index(inplace = True)
```

```python
df1.head()
df1.tail()
```

```python
df1.to_csv("F:\\onedrive\\Desktop\\DATA SCIENCE\\Project 1\\Cross-Market Analysis\\Cross-Market Data\\raw\\stocks\\gspc_data.csv")
```
