# Final Model Comparison

This notebook consolidates the regression result already created earlier with the clustering model selected in the clustering phase.

```python
import yaml
from src.utils.io import read_csv_file
from src.utils.paths import find_project_root, resolve_project_path
```

```python
root = find_project_root()

regression_comparison_path = resolve_project_path(root, "artifacts/comparison/model_comparison_table.csv")
clustering_results_path = resolve_project_path(root, "artifacts/clustering/silhouette_results.csv")
final_selected_models_path = resolve_project_path(root, "artifacts/comparison/final_selected_models.yaml")
```

```python
regression_df = read_csv_file(regression_comparison_path)
clustering_df = read_csv_file(clustering_results_path)

regression_df.head(), clustering_df.head()
```

```python
best_regression_model = regression_df.iloc[0]["model_name"]
best_clustering_model = clustering_df.iloc[0]["model_name"]

final_selected_models = {
    "best_regression_model": best_regression_model,
    "best_clustering_model": best_clustering_model,
}
final_selected_models
```

```python
with open(final_selected_models_path, "w", encoding="utf-8") as f:
    yaml.safe_dump(final_selected_models, f, sort_keys=False, allow_unicode=True)

final_selected_models_path
```
